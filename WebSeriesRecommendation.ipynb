```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Web Series Dataset.csv")
```

```python
# to drop null values
df.dropna(inplace=True)
```

```python
df.head()
```

```python
# to remove no.of seasons column
df.drop('No of Seasons', axis=1, inplace=True)
```

```python
df.head()
```

```python
import re
```

```python
import string

# text preprocessing on description column
def clean_text(text):
    if pd.isnull(text):
        return ""
    text = text.lower()  # convert to lowercase
    text = re.sub(r'\d+', '', text)  # remove digits
    text = re.sub(rf'[{re.escape(string.punctuation)}]', '', text)  # remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra whitespace
    return text
```

```python
df['Cleaned_Description'] = df['Description'].apply(clean_text)
df['Cleaned_Genre'] = df['Genre'].apply(clean_text)
```

```python
# Combine Description and Genre into a single feature
df['Combined_Features'] = df['Cleaned_Description'] + ' ' + df['Cleaned_Genre']
```

```python
# Display cleaned and combined text columns
df[['Description', 'Genre', 'Cleaned_Description', 'Cleaned_Genre', 'Combined_Features']].head()
```

```python
# to remove description and genre columns
df.drop(['Description', 'Genre'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# to remove streaming platform column
df.drop('Streaming Platform', axis=1, inplace=True)
```

```python
df.head()
```

```python
# to remove cleaned description column and cleaned genre column
df.drop(['Cleaned_Description', 'Cleaned_Genre'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# to do count vectorization
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Combined_Features'])

# to do cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(X)
```

```python
# to train the model based on combined features
def get_recommendations(series):
        series_index = df[df['Series Title'] == series].index[0]
        distances = sorted(list(enumerate(similarity[series_index])), reverse=True, key=lambda x: x[1])
        recommended_series = []
        for i in distances[1:6]:
            recommended_series.append(df.iloc[i[0]]['Series Title'])
        return recommended_series
```

```python
recommendations = get_recommendations('Breaking Bad')
print(recommendations)
```

```python
recommendations = get_recommendations('Game of Thrones')
print(recommendations)
```
